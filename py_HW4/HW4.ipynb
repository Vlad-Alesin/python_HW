#### 1.2. Изучите стоимости недвижимости

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kc_house_data.csv')
df.head()
```

```python
plt.hist(df['price'], bins=100)

plt.title('Стоимость жилья')
plt.xlabel('цена')
plt.ylabel('количество')
```

Основное распределение цены в диапазоне примерно от 300 до 700 тыс. долларов

#### 1.3. Изучите распределение квадратуры жилой

```python
plt.hist(df['sqft_living'], bins=50)
plt.title('Распределение жилой площади в кв.м')
plt.xlabel('Площадь')
plt.ylabel('Количество')
```

Больше всего домов с квадратурой от 1000 до 3000 м.кв.

#### 1.4. Изучите распределение года постройки

```python
plt.hist(df['yr_built'], bins=50)
plt.title('Распределение по году постройки')
plt.xlabel('Год постройки дома')
plt.ylabel('Количество')
```

#### 2.1. Изучите распределение домов от наличия вида на набережную

#### 2.2. Изучите распределение этажей домов

```python
data = df['floors'].value_counts()
data

plt.figure(figsize=(5, 3))
plt.bar(data.index, data.values, color='#CCFF99', edgecolor='#183300', width=0.4)
plt.title('Распределение этажности домов')
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов')
```

#### 2.3. Изучите распределение состояния домов

```python
data = df['condition'].value_counts()
data
plt.figure(figsize=(6, 4))
plt.bar(data.index, data.values, color='#CCFF99', edgecolor='#183300', width=0.8)
plt.title('Распределение домов в зависимости от их состояния')
plt.xlabel('Оценка состояния')
plt.ylabel('Количество домов')
```

#### Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.jointplot(x=df['price'], y=df['bedrooms'], kind='reg')
```

На графике видно, что наибольшее кол-во зданий продаются с 3-6 комнатами и цена практически не зависит от их кол-ва

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg')
```

Из графика видно, что наибольшее кол-во домов имеет жилую площадь от 1 до 5 тысяч квадратных футов (если я правильно понял единицы  измерения)  

```python
sns.jointplot(x=df['price'], y=df['lat'], kind='reg')
```

По графику видно большое скопление домов на широте 47,6. Так же видно увеличение кол-ва более дорогих домов на тех же широтах

```python
sns.jointplot(x=df['price'], y=df['yr_built'], kind='reg')
```

На графике видно, что цены практически не зависят от года постройки (однако более новых домов больше, чем старых) 

```python
sns.jointplot(x=df['price'], y=df['waterfront'], kind='reg')
```

Анализ показал, что домов с видом на набережную гораздо меньше, чем домов без вида на набережную. В то же время видно, что наличие вида на набережную не влияет на стоимость

```python
corr_matrix = df[['price', 'bedrooms', 'bathrooms', 'sqft_lot', 'floors', 'view', 'condition', 'grade', 'sqft_basement']].corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')
```

Видно, что наибольшая зависимость цены проявляется от класса здания.
Так же наблюдается зависимость цены от количества ванных комнат и вида из дома. Кроме того наблюдается малая зависимость от этажности здания, кол-ва комнат и площади подвала и отсутствует зависимость цены от общей площади здания и его состояния
