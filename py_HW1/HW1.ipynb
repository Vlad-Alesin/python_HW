## Задача 3:

Дан список с затратами на рекламу. Но в данных есть ошибки, некоторые затраты имеют отрицательную величину. Удалите такие значения из списка и
посчитайте суммарные затраты
[100, 125, -90, 345, 655, -1, 0, 200]
Используйте list comprehensions

```python
expenses = [100, 125, -90, 345, 655, -1, 0, 200]
rightExpenses = [i if i > 0 else 0 for i in expenses]
sum = 0
for j in range (0, len(rightExpenses)):
  sum += rightExpenses[j]
print(sum)
```
